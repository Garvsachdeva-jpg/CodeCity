```python
import os
import sqlite3

import pandas as pd

BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))
DB_PATH = os.path.join(BASE_DIR, "codecity.db")

conn = sqlite3.connect(DB_PATH)

query = """
SELECT
  snapshot_id,
  path,
  name,
  size,
  complexity,
  width,
  depth,
  height,
  extension,
  is_test_file,
  area,
  aspect_ratio
FROM files
"""

files_df = pd.read_sql_query(query, conn)
conn.close()

print(files_df.head())
print("Rows:", len(files_df))
print(files_df.describe(include="all"))
```
